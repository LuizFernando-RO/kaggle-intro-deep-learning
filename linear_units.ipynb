Next, we create the network architecture. In this first tutorial, we are using a single neuron (which works like a linear regression) to predict the number of calories of a cereal given 3 features. Hence, the input has shape 3 and the output is a single value.

To create the 

```python
model = keras.Sequential([
    layers.Dense(units=1, input_shape=[3])
])
```

We can check the weights of the model by accessing the _.weights_ attribute. It returns us both the weights and the bias. Since out model was not trained yet, the weights are randomly initialized and the bias is set to 0.

```python
w, b = model.weights
print(f"Weights\n{w}\n\nBias\n{b}")
```
